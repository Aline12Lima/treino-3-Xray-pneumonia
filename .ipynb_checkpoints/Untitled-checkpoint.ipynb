```python
# --- Bibliotecas ---
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Conv2D, Multiply, Add, Activation, Concatenate, Lambda, GlobalMaxPooling2D, Reshape
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from collections import Counter
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import random
import os

# --- Reprodutibilidade ---
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# --- Diretórios ---
DATASET_PATH = "/home/aline/projetos/Pesquisa_Biologia-teste2/dataset/chest_xray"
train_dir = "dataset/chest_xray/train"
test_dir = "dataset/chest_xray/test" 

# --- Data Augmentation ---
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

# --- Geradores de Dados ---
print("--- Preparando Geradores de Dados ---")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="training"
)
val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="validation"
)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    shuffle=False
)
print("\nGeradores de dados prontos.")
```

```python
print("Definindo funções do CBAM...")

# --- FUNÇÕES DO CBAM (CORRIGIDAS) ---
def channel_attention_module(x, ratio=8):
    channel = x.shape[-1]
    avg_pool = GlobalAveragePooling2D()(x)
    avg_pool = Reshape((1, 1, channel))(avg_pool)
    max_pool = GlobalMaxPooling2D()(x)
    max_pool = Reshape((1, 1, channel))(max_pool)
    shared_mlp_one = Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_mlp_two = Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_mlp_two(shared_mlp_one(avg_pool))
    max_out = shared_mlp_two(shared_mlp_one(max_pool))
    attention = Add()([avg_out, max_out])
    attention = Activation('sigmoid')(attention)
    return Multiply()([x, attention])

def spatial_attention_module(x):
    avg_pool = Lambda(lambda t: tf.reduce_mean(t, axis=3, keepdims=True))(x)
    max_pool = Lambda(lambda t: tf.reduce_max(t, axis=3, keepdims=True))(x)
    concat = Concatenate(axis=3)([avg_pool, max_pool])
    attention = Conv2D(1, kernel_size=7, padding='same', activation='sigmoid', kernel_initializer='he_normal', use_bias=False)(concat)
    return Multiply()([x, attention])

def cbam_block(x, ratio=8):
    x = channel_attention_module(x, ratio)
    x = spatial_attention_module(x)
    return x

print("Funções do CBAM prontas.")
print("\nConstruindo o modelo...")

# --- CONSTRUÇÃO DO MODELO ---
base_model = DenseNet121(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
x = base_model.output 
x = cbam_block(x) 
print("Bloco CBAM inserido.")
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation="sigmoid")(x)
model = Model(inputs=base_model.input, outputs=predictions)
print("Modelo construído.")

# --- COMPILAÇÃO E BALANCEAMENTO ---
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss="binary_crossentropy",
              metrics=["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])
print("Modelo compilado.")
counter = Counter(train_generator.classes)
max_val = float(max(counter.values()))
class_weights = {class_id: max_val/num_images for class_id, num_images in counter.items()}
print("Pesos de classe calculados:", class_weights)

# --- SUMÁRIO FINAL ---
print("\n--- SUMÁRIO DO MODELO CONSTRUÍDO ---")
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# --- 1. DEFINA OS CALLBACKS ---
checkpoint = ModelCheckpoint(
    filepath='melhor_modelo_teste2.keras',
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    mode='min',
    restore_best_weights=True
)

# --- 2. EXECUTE O TREINAMENTO COM OS CALLBACKS ---
print("\nIniciando Treinamento com Callbacks (15 Épocas)...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,
    class_weight=class_weights,
    callbacks=[checkpoint, early_stopping]
)

print("\nTreinamento concluído.")

# --- AVALIAÇÃO FINAL ---
print("\n--- Avaliação no Conjunto de Teste (com os melhores pesos restaurados) ---")
test_loss, test_acc, test_prec, test_rec, test_auc = model.evaluate(test_generator)
print(f"Teste - Loss: {test_loss:.4f}, Acc: {test_acc:.4f}, Precision: {test_prec:.4f}, Recall: {test_rec:.4f}, AUC: {test_auc:.4f}")

# --- GRÁFICOS ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Acurácia de Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia do Modelo')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perda de Treino')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda do Modelo')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# --- 1. Resetar o gerador de teste ---
test_generator.reset()

# --- 2. Fazer previsões ---
predictions = model.predict(test_generator, verbose=1)
y_pred_classes = (predictions.flatten() > 0.5).astype(int)  # binariza
y_true = test_generator.classes
file_names = test_generator.filenames

# --- 3. Criar DataFrame com resultados ---
import pandas as pd

df_results = pd.DataFrame({
    'FileName': file_names,
    'Y_True': y_true,
    'Y_Pred': y_pred_classes,
    'Prob_Pneumonia': predictions.flatten()
})

# --- 4. Filtrar Falsos Negativos e Falsos Positivos ---
df_erros_fn = df_results[(df_results['Y_True'] == 1) & (df_results['Y_Pred'] == 0)].copy()
df_erros_fp = df_results[(df_results['Y_True'] == 0) & (df_results['Y_Pred'] == 1)].copy()

# --- 5. Mostrar quantidade de erros ---
print(f"Falsos Negativos (FN): {len(df_erros_fn)}")
print(f"Falsos Positivos (FP): {len(df_erros_fp)}")

fp_files = df_erros_fp['FileName'].tolist()
fn_files = df_erros_fn['FileName'].tolist()
#printes dos arquivos
print("Arquivos Falsos Positivos (FP):")
for f in df_erros_fp['FileName']:
    print(f"- {f}")
print("\nArquivos Falsos Negativos (FN):")
for f in df_erros_fn['FileName']:
    print(f"- {f}")
df_erros_fp[['FileName']].to_csv("FP_files.csv", index=False)
df_erros_fn[['FileName']].to_csv("FN_files.csv", index=False)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# --- 1. Combinar FP e FN para análise geral ---
df_errors = pd.concat([df_erros_fp, df_erros_fn], ignore_index=True)

# --- 2. Selecionar features numéricas para clusterização ---
# Aqui usamos probabilidade prevista e classe verdadeira como exemplo
X = df_errors[['Prob_Pneumonia', 'Y_True']].values

# --- 3. Normalizar features (importante para K-Means) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Definir número de clusters (k) ---
k = 3  # ajuste conforme quiser
kmeans = KMeans(n_clusters=k, random_state=42)
df_errors['Cluster'] = kmeans.fit_predict(X_scaled)

# --- 5. Visualizar resultados (scatter plot simples) ---
plt.figure(figsize=(8,6))
for c in range(k):
    cluster_points = X_scaled[df_errors['Cluster'] == c]
    plt.scatter(cluster_points[:,0], cluster_points[:,1], label=f'Cluster {c}')
plt.xlabel('Probabilidade (normalizada)')
plt.ylabel('Classe Verdadeira (normalizada)')
plt.title('Clusterização de erros FP+FN')
plt.legend()
plt.show()

# --- 6. Estatísticas por cluster ---
for c in range(k):
    cluster_df = df_errors[df_errors['Cluster'] == c]
    print(f"\nCluster {c}:")
    print(f"Quantidade de erros: {len(cluster_df)}")
    print(f"Probabilidade média: {cluster_df['Prob_Pneumonia'].mean():.3f}")
    print(f"Distribuição por classe verdadeira:\n{cluster_df['Y_True'].value_counts()}")
```

```python
# VISUALIZAÇÃO - MAPA CBAM -  FN----
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
from tensorflow.keras.models import Model

SEED = 42
random.seed(SEED)

# Define o tamanho da imagem que o modelo espera
IMG_SIZE = (224, 224)

def get_cbam_maps(model, img_tensor):
    """Extrai mapas de atenção CBAM (channel + spatial) do modelo para uma imagem"""
    
    cbam_layers = [l for l in model.layers if 'multiply' in l.name.lower()]
    if not cbam_layers:
        print("CBAM layers não encontradas!")
        return None
    
    cbam_model = Model(inputs=model.input, outputs=[l.output for l in cbam_layers])
    attention_maps = cbam_model.predict(img_tensor[np.newaxis, ...])
    return attention_maps

# Seleciona 5 imagens aleatórias de Falsos Negativos
indices_fn = random.sample(range(len(df_erros_fn)), min(5, len(df_erros_fn)))

for idx in indices_fn:
    file_path = os.path.join(test_dir, df_erros_fn.iloc[idx]['FileName'])
    img = cv2.imread(file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img, IMG_SIZE)
    img_resized = img_resized / 255.0

    att_maps = get_cbam_maps(model, img_resized)

    plt.figure(figsize=(15, 5))
    plt.subplot(1, len(att_maps)+1, 1)
    plt.imshow(img_resized)
    plt.title("Original FN")
    plt.axis('off')

    for i, att in enumerate(att_maps):
        plt.subplot(1, len(att_maps)+1, i+2)
        plt.imshow(att[0, :, :, 0], cmap='jet')
        plt.title(f"CBAM {i+1}")
        plt.axis('off')
    plt.show()
    
```

```python
# VISUALIZAÇÃO MAPA - CBAM - FP 
indices_fp = random.sample(range(len(df_erros_fp)), min(5, len(df_erros_fp)))

for idx in indices_fp:
    file_path = os.path.join(test_dir, df_erros_fp.iloc[idx]['FileName'])
    img = cv2.imread(file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img, IMG_SIZE)
    img_resized = img_resized / 255.0

    att_maps = get_cbam_maps(model, img_resized)

    plt.figure(figsize=(15, 5))
    plt.subplot(1, len(att_maps)+1, 1)
    plt.imshow(img_resized)
    plt.title("Original FP")
    plt.axis('off')

    for i, att in enumerate(att_maps):
        plt.subplot(1, len(att_maps)+1, i+2)
        plt.imshow(att[0, :, :, 0], cmap='jet')
        plt.title(f"CBAM {i+1}")
        plt.axis('off')
    plt.show()
```
