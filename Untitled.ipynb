```python
# --- Célula 1 -- Configuração inicial e importações de bibliotecas ------
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import (Dense, GlobalAveragePooling2D, GlobalMaxPooling2D,
                                     Dropout, Conv2D, Multiply, Add, Activation,
                                     Concatenate, Reshape)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from collections import Counter
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import random
import os
import cv2  # <-- Importação crucial que adicionamos para as visualizações

# --- Reprodutibilidade ---
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# --- Diretórios ---
train_dir = "dataset/chest_xray/train"
test_dir = "dataset/chest_xray/test"

# --- Data Augmentation ---
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

# --- Geradores de Dados ---
print("--- Preparando Geradores de Dados ---")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="training"
)
val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="validation"
)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    shuffle=False
)
print("\nGeradores de dados prontos.")
```

```python
# --- Célula 2  ---
print("Definindo a arquitetura do modelo...")

import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Dense, GlobalAveragePooling2D, Dropout, Conv2D,
                                     Multiply, Add, Activation, Concatenate, Reshape,
                                     GlobalMaxPooling2D, Layer)
from collections import Counter

# --- CAMADAS CUSTOMIZADAS CBAM ---

def channel_attention_module(x, ratio=8):
    """Cria o módulo de atenção de canal."""
    channel = x.shape[-1]
    shared_mlp_one = Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_mlp_two = Dense(channel, kernel_initializer='he_normal', use_bias=False)
    
    avg_pool = GlobalAveragePooling2D()(x)
    avg_pool = Reshape((1, 1, channel))(avg_pool)
    avg_out = shared_mlp_two(shared_mlp_one(avg_pool))
    
    max_pool = GlobalMaxPooling2D()(x)
    max_pool = Reshape((1, 1, channel))(max_pool)
    max_out = shared_mlp_two(shared_mlp_one(max_pool))
    
    attention = Activation('sigmoid')(Add()([avg_out, max_out]))
    # Nomeamos a camada de saída para encontrá-la facilmente depois
    return Multiply(name='channel_attention_output')([x, attention])


class SpatialAttention(Layer):
    """Cria o módulo de atenção espacial como uma Camada Customizada."""
    def __init__(self, kernel_size=7, **kwargs):
        super(SpatialAttention, self).__init__(**kwargs)
        self.conv2d = Conv2D(filters=1, kernel_size=kernel_size, padding='same',
                             activation='sigmoid', kernel_initializer='he_normal',
                             use_bias=False)

    def call(self, inputs):
        avg_pool = tf.reduce_mean(inputs, axis=3, keepdims=True)
        max_pool = tf.reduce_max(inputs, axis=3, keepdims=True)
        concat = Concatenate(axis=3)([avg_pool, max_pool])
        attention = self.conv2d(concat)
       
        return Multiply(name='spatial_attention_output')([inputs, attention])

def cbam_block(x, ratio=8):
    """Combina os dois módulos de atenção."""
    x = channel_attention_module(x, ratio)
    x = SpatialAttention(name='spatial_attention')(x)
    return x

print("Funções e classes do CBAM prontas.")

# --- CONSTRUÇÃO DO MODELO ---
print("\nConstruindo o modelo...")
base_model = DenseNet121(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = cbam_block(x) # Insere o bloco CBAM corrigido
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation="sigmoid")(x)

model = Model(inputs=base_model.input, outputs=predictions)
print("Modelo construído.")

# --- COMPILAÇÃO ---
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss="binary_crossentropy",
    metrics=["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()]
)
print("Modelo compilado.")

# --- PESOS DE CLASSE ---
try:
    counter = Counter(train_generator.classes)
    max_val = float(max(counter.values()))
    class_weights = {class_id: max_val / num_images for class_id, num_images in counter.items()}
    print("Pesos de classe calculados:", class_weights)
except NameError:
    print("Aviso: 'train_generator' não definido. Rode a Célula 1 primeiro.")

print("\n--- SUMÁRIO DO MODELO ---")
model.summary()
```

```python
# --- celula 3 -- -Treino com  CALLBACKS ---
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping


checkpoint = ModelCheckpoint(
    filepath='melhor_modelo_teste2.keras',
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    mode='min',
    restore_best_weights=True
)

# --- 2. EXECUTE O TREINAMENTO COM OS CALLBACKS ---
print("\nIniciando Treinamento com Callbacks (15 Épocas)...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,
    class_weight=class_weights,
    callbacks=[checkpoint, early_stopping]
)

print("\nTreinamento concluído.")

# --- AVALIAÇÃO FINAL ---
print("\n--- Avaliação no Conjunto de Teste (com os melhores pesos restaurados) ---")
test_loss, test_acc, test_prec, test_rec, test_auc = model.evaluate(test_generator)
print(f"Teste - Loss: {test_loss:.4f}, Acc: {test_acc:.4f}, Precision: {test_prec:.4f}, Recall: {test_rec:.4f}, AUC: {test_auc:.4f}")

# --- GRÁFICOS ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Acurácia de Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia do Modelo')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perda de Treino')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda do Modelo')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# celula 4 - coletas de FP e FN ---- 

base_model = DenseNet121(weights="imagenet", include_top=False, input_shape=(224, 224, 3))

base_model.trainable = False

x = base_model.output

x = cbam_block(x)

x = GlobalAveragePooling2D()(x)

x = Dropout(0.5)(x)

predictions = Dense(1, activation="sigmoid")(x)

model = Model(inputs=base_model.input, outputs=predictions)



# 2. Carregar pesos salvos

model.load_weights("melhor_modelo_teste2.keras")



# 3. Resetar gerador de teste e prever

test_generator.reset()

predictions = model.predict(test_generator)

y_pred_classes = np.where(predictions.flatten() > 0.5, 1, 0)

y_true = test_generator.classes

file_names = test_generator.filenames



# 4. Criar DataFrame com resultados

df_results = pd.DataFrame({

    'FileName': file_names,

    'Y_True': y_true,

    'Y_Pred': y_pred_classes,

    'Prob_Pneumonia': predictions.flatten()

})



# 5. Filtrar erros

df_erros_fn = df_results[(df_results['Y_True'] == 1) & (df_results['Y_Pred'] == 0)].copy()

df_erros_fp = df_results[(df_results['Y_True'] == 0) & (df_results['Y_Pred'] == 1)].copy()



print(f"\nAnálise de erros concluída.")

print(f"Total de Falsos Negativos (FN) encontrados: {len(df_erros_fn)}")

print(f"Total de Falsos Positivos (FP) encontrados: {len(df_erros_fp)}")



```

```python
# celula 5 aplica clusterização sobre a coleta ----- 
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# --- 1. Combinar FP e FN para análise geral ---
df_errors = pd.concat([df_erros_fp, df_erros_fn], ignore_index=True)

# --- 2. Selecionar features numéricas para clusterização ---
# Aqui usamos probabilidade prevista e classe verdadeira como exemplo
X = df_errors[['Prob_Pneumonia', 'Y_True']].values

# --- 3. Normalizar features (importante para K-Means) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Definir número de clusters (k) ---
k = 3  # ajuste conforme quiser
kmeans = KMeans(n_clusters=k, random_state=42)
df_errors['Cluster'] = kmeans.fit_predict(X_scaled)

# --- 5. Visualizar resultados (scatter plot simples) ---
plt.figure(figsize=(8,6))
for c in range(k):
    cluster_points = X_scaled[df_errors['Cluster'] == c]
    plt.scatter(cluster_points[:,0], cluster_points[:,1], label=f'Cluster {c}')
plt.xlabel('Probabilidade (normalizada)')
plt.ylabel('Classe Verdadeira (normalizada)')
plt.title('Clusterização de erros FP+FN')
plt.legend()
plt.show()

# --- 6. Estatísticas por cluster ---
for c in range(k):
    cluster_df = df_errors[df_errors['Cluster'] == c]
    print(f"\nCluster {c}:")
    print(f"Quantidade de erros: {len(cluster_df)}")
    print(f"Probabilidade média: {cluster_df['Prob_Pneumonia'].mean():.3f}")
    print(f"Distribuição por classe verdadeira:\n{cluster_df['Y_True'].value_counts()}")
```

```python
# --- Célula 6 (Unificada e Robusta) --- Visualização dos Mapas de Atenção ---

# Parte 1: Função para extrair os mapas por NOME
def get_cbam_maps(model, img_tensor):
    """Extrai mapas de atenção CBAM por nome de camada, de forma robusta."""
    try:
        # Pede as camadas pelos nomes exatos que definimos na Célula 2
        channel_out_layer = model.get_layer('channel_attention_output')
        spatial_out_layer = model.get_layer('spatial_attention')
        
        # Cria o sub-modelo para expor as saídas dessas camadas
        cbam_model = Model(inputs=model.input, outputs=[channel_out_layer.output, spatial_out_layer.output])
        attention_maps = cbam_model.predict(img_tensor[np.newaxis, ...], verbose=0) # verbose=0 para uma saída mais limpa
        return attention_maps
    except ValueError as e:
        print(f"ERRO: Não foi possível encontrar as camadas de atenção. Verifique os nomes na Célula 2. Detalhe: {e}")
        return None

# Parte 2: Função reutilizável para plotar
def plotar_mapas_de_atencao(df_erros, tipo_erro_str, num_exemplos=5):
    """Seleciona uma amostra de erros de um DataFrame, gera e plota os mapas de atenção CBAM."""
    print(f"\n--- Visualizando {num_exemplos} exemplos aleatórios de {tipo_erro_str} ---")
    
    num_a_mostrar = min(num_exemplos, len(df_erros))
    if num_a_mostrar == 0:
        print(f"Nenhum erro do tipo {tipo_erro_str} para visualizar.")
        return
        
    indices = random.sample(range(len(df_erros)), num_a_mostrar)

    for idx in indices:
        file_path = os.path.join(test_dir, df_erros.iloc[idx]['FileName'])
        
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_resized = cv2.resize(img, (224, 224)) / 255.0

        att_maps = get_cbam_maps(model, img_resized)

        if att_maps:
            plt.figure(figsize=(15, 5))
            plt.subplot(1, len(att_maps) + 1, 1)
            plt.imshow(img_resized)
            plt.title(f"Original {tipo_erro_str}")
            plt.axis('off')

            for i, att in enumerate(att_maps):
                plt.subplot(1, len(att_maps) + 1, i + 2)
                heatmap = np.mean(att[0], axis=-1)
                plt.imshow(heatmap, cmap='jet')
                plt.title(f"CBAM Map {i+1}")
                plt.axis('off')
            plt.show()

# Parte 3: Chamadas da função
# Simplesmente chamamos nossa função para cada grupo de erro
plotar_mapas_de_atencao(df_erros_fn, "FN")
plotar_mapas_de_atencao(df_erros_fp, "FP")
```

```python
# --- celula 7- Visão qualitativa ---- FUNÇÃO AUXILIAR PARA PLOTAR (para não repetir código) ---
def plotar_exemplos_cluster(df_cluster, cluster_id, tipo_erro, num_exemplos=3):
    """
    Seleciona imagens de um cluster específico e plota suas ativações CBAM.
    """
    print(f"\n--- Visualizando Exemplos do Cluster {cluster_id} ({tipo_erro}) ---")
    
    # Seleciona 'num_exemplos' aleatórios do dataframe do cluster
    amostra_df = df_cluster.sample(min(num_exemplos, len(df_cluster)), random_state=SEED)
    
    if amostra_df.empty:
        print(f"Cluster {cluster_id} não tem amostras para exibir.")
        return

    for _, row in amostra_df.iterrows():
        file_path = os.path.join(test_dir, row['FileName'])
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_resized = cv2.resize(img, (224, 224)) / 255.0

        # A função 'get_cbam_maps' foi definida na Célula 6
        att_maps = get_cbam_maps(model, img_resized)

        if att_maps:
            plt.figure(figsize=(12, 4))
            plt.subplot(1, len(att_maps) + 1, 1)
            plt.imshow(img_resized)
            plt.title(f"Original ({tipo_erro} - Cluster {cluster_id})\nProb: {row['Prob_Pneumonia']:.2f}")
            plt.axis('off')

            for i, att in enumerate(att_maps):
                plt.subplot(1, len(att_maps) + 1, i + 2)
                # O mapa de atenção é a média dos canais de saída
                heatmap = np.mean(att[0], axis=-1)
                plt.imshow(heatmap, cmap='jet')
                plt.title(f"CBAM Map {i+1}")
                plt.axis('off')
            plt.show()

# --- GERAR VISUALIZAÇÕES PARA CADA CLUSTER DE ERRO ---


plotar_exemplos_cluster(df_errors[df_errors['Cluster'] == 1], cluster_id=1, tipo_erro="FN - Ponto Cego")
plotar_exemplos_cluster(df_errors[df_errors['Cluster'] == 2], cluster_id=2, tipo_erro="FP - Alta Confiança")
plotar_exemplos_cluster(df_errors[df_errors['Cluster'] == 0], cluster_id=0, tipo_erro="FP - Baixa Confiança")
```

```python
# --- CÉLULA 9 --- visão quantitativa ----ANÁLISE DOS CENTRÓIDES DOS CLUSTERS DE ERRO ---

import pandas as pd
import numpy as np

print("--- Análise dos Centróides dos Clusters de Erro ---")

# 1. ACESSAR OS CENTRÓIDES E DADOS NECESSÁRIOS
# O objeto 'kmeans' e o 'df_errors' devem vir da Célula 5
try:
    centroids = kmeans.cluster_centers_
    print(f"Centróides extraídos com sucesso. Formato: {centroids.shape}")
except NameError:
    print("ERRO: Objeto 'kmeans' não encontrado. Rode a Célula 5 primeiro.")
    # Interrompe a execução se a célula anterior não foi rodada
    raise

# 2. ANÁLISE QUANTITATIVA
# Calcula o centróide médio para servir como uma linha de base
mean_centroid = np.mean(centroids, axis=0)

print("\n--- ANÁLISE DOS FEATURES MAIS DISTINTIVOS POR CLUSTER ---")

# Itera sobre cada cluster que você encontrou (0, 1, 2)
for i in range(kmeans.n_clusters):
    
    # Pega os dados do cluster atual
    cluster_df = df_errors[df_errors['Cluster'] == i]
    centroid = centroids[i]
    
    # Calcula a diferença para encontrar os features mais únicos de cada cluster
    diff = np.abs(centroid - mean_centroid)
    # Pega os índices dos 10 features com a maior diferença
    top_10_indices = np.argsort(diff)[-10:][::-1]
    
    # Pega a estatística de probabilidade que já tínhamos
    prob_media = cluster_df['Prob_Pneumonia'].mean()
    tipo_erro_predominante = "FN" if 1 in cluster_df['Y_True'].unique() else "FP"
    
    print(f"\nCluster {i}: ({len(cluster_df)} erros - Tipo Predominante: {tipo_erro_predominante})")
    print(f"Confiança Média no Erro (Prob. Pneumonia): {prob_media:.4f}")
    print(f"As features da clusterização que mais definem este cluster:")
    
    # Cria um DataFrame para visualização clara dos resultados
    analysis_df = pd.DataFrame({
        'Indice_Feature': top_10_indices,
        'Valor_no_Centroide': centroid[top_10_indices],
        'Importancia_Relativa': diff[top_10_indices]
    })
    
    print(analysis_df)
```
